```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Load preprocessed data
df = pd.read_csv(r"C:\Users\Musae\Documents\GitHub-REPOs\Crop-Classifier-From-Space\preprocessing\processed_crop_dataset.csv")

```

```python
df
```

```python

# Feature columns
features = ['B2', 'B3', 'B4', 'B8', 'NDVI', 'GNDVI', 'EVI']

# Encode target labels
df['crop'] = df['crop'].astype('category')
df['crop_label'] = df['crop'].cat.codes  # cocoa=0, oil_palm=1, rubber=2 (check df['crop'].cat.categories)

# Train-test split
X = df[features]
y = df['crop_label']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Train classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Optional: Show label mapping
print("\nLabel Mapping:", dict(enumerate(df['crop'].cat.categories)))
```
